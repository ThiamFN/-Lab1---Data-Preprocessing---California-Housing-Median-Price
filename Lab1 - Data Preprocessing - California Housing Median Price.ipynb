### Identifier les features et le Label

- Combien de features/ caractéristiques sont présentes dans le jeu de données?
- Quels sont leurs types?
- Le Label est-il continu ou catégoriel?

```python
def analyser_dataframe(df, label_column):
    # Nombre de caractéristiques (excluant la colonne du label)
    num_features = len(df.columns) - 1

    # Types des caractéristiques
    feature_types = {col: df[col].dtype for col in df.columns if col != label_column}

    # Détection du type de label
    if pd.api.types.is_numeric_dtype(df[label_column]):
            label_type = 'Continu'
    else:
        label_type = 'Catégoriel'

    return num_features, feature_types, label_type
```

```python
label_column = 'median_house_value'
num_features, feature_types, label_type = analyser_dataframe(housing_df, label_column)

print(f"Nombre de caractéristiques : {num_features}")
print("Types des caractéristiques :")
for col, col_type in feature_types.items():
    print(f"  - {col}: {col_type}")
print(f"Le label est de type : {label_type}")
```
